```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data from your results
data = {
    'Filter': [
        'Before Filter',
        'All Voting',
        'OR Voting',
        'Major Voting',
        'KNN Filter',
        'Decision Tree Filter',
        'Random Forest Filter'
    ],
    'RF-100': [0.8786, 0.8352, 0.8784, 0.8785, 0.8775, 0.8777, 0.8787],
    'RF-250': [0.8796, 0.8366, 0.8794, 0.8795, 0.8788, 0.8785, 0.8803],
    'RF-500': [0.8800, 0.8369, 0.8805, 0.8797, 0.8787, 0.8789, 0.8806],
    'Linear Reg': [0.8881, 0.8912, 0.8885, 0.8912, 0.8912, 0.8900, 0.8912],
    'KNN': [0.8742, 0.8821, 0.8727, 0.8724, 0.8703, 0.8716, 0.8701],
    'Decision Tree': [0.8279, 0.8059, 0.8263, 0.8290, 0.8261, 0.8267, 0.8159]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set up the plot
fig, ax = plt.subplots(figsize=(15, 8))

# Define colors for each model
colors = {
    'RF-100': '#8884d8',
    'RF-250': '#82ca9d', 
    'RF-500': '#ffc658',
    'Linear Reg': '#ff7300',
    'KNN': '#00ff88',
    'Decision Tree': '#ff0088'
}

# Set width of bars and positions
n_groups = len(df)
n_bars = 6
bar_width = 0.12
index = np.arange(n_groups)

# Create bars for each model
bars = []
model_names = ['RF-100', 'RF-250', 'RF-500', 'Linear Reg', 'KNN', 'Decision Tree']
display_names = ['RF (100 trees)', 'RF (250 trees)', 'RF (500 trees)', 
                'Linear Regression', 'K-Nearest Neighbors', 'Decision Tree']

for i, (model, display_name) in enumerate(zip(model_names, display_names)):
    bar = ax.bar(index + i * bar_width, df[model], bar_width, 
                label=display_name, color=colors[model], alpha=0.8)
    bars.append(bar)

# Customize the plot
ax.set_xlabel('Filter Type', fontsize=12, fontweight='bold')
ax.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax.set_title('Machine Learning Model Accuracy Comparison Across Different Filters', 
             fontsize=14, fontweight='bold', pad=20)

# Set x-axis labels
ax.set_xticks(index + bar_width * (n_bars-1) / 2)
ax.set_xticklabels(df['Filter'], rotation=45, ha='right')

# Set y-axis limits for better visualization
ax.set_ylim(0.80, 0.90)

# Add grid for better readability
ax.grid(True, alpha=0.3, axis='y')

# Add legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Add value labels on bars (optional - uncomment if needed)
# for bar_group in bars:
#     for bar in bar_group:
#         height = bar.get_height()
#         ax.text(bar.get_x() + bar.get_width()/2., height + 0.001,
#                f'{height:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

# Create a summary statistics table
print("\nSummary Statistics:")
print("="*80)
summary_stats = df.set_index('Filter').describe()
print(summary_stats.round(4))

# Find best performing model for each filter
print("\nBest Performing Model for Each Filter:")
print("="*80)
for i, filter_name in enumerate(df['Filter']):
    row = df.iloc[i, 1:]  # Exclude the 'Filter' column
    best_model = row.idxmax()
    best_accuracy = row.max()
    print(f"{filter_name:25}: {best_model:12} (Accuracy: {best_accuracy:.4f})")

# Find most consistent model across filters
print("\nModel Consistency (Standard Deviation across filters):")
print("="*80)
consistency = df.set_index('Filter').std().sort_values()
for model, std_dev in consistency.items():
    print(f"{model:15}: {std_dev:.4f}")

# Alternative visualization: Grouped bar chart by model
fig2, ax2 = plt.subplots(figsize=(12, 8))

# Transpose data for model-centric view
df_transposed = df.set_index('Filter').T

# Create grouped bar chart
df_transposed.plot(kind='bar', ax=ax2, width=0.8, 
                  color=[colors[col] if col in colors else '#cccccc' for col in df_transposed.columns])

ax2.set_title('Model Performance Across Different Filters', fontsize=14, fontweight='bold')
ax2.set_xlabel('Machine Learning Models', fontsize=12, fontweight='bold')
ax2.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax2.set_ylim(0.80, 0.90)
ax2.grid(True, alpha=0.3, axis='y')
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Filters')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
